```python
import json
import re

with open('rozvrh.json', encoding='utf-8') as inputFile:
    data = json.load(inputFile)
```

```python
sourceTable = []

for event in data['events']:
    teachersNames = event.get ("teachersNames", [])
    teachersCount = len(teachersNames)
    if not "lessonsCount" in event:
        continue
    for teachers in teachersNames:
        teacher_name = teachers.split(' (')[0].strip().replace(',', '')
        row = {}
        row["lessonsCount"] = event["lessonsCount"] /teachersCount
        row["lessonFormName"] = event["lessonFormName"]
        row["teachersIds"] = event["teachersIds"]
        row["teachers"] = teachers
        sourceTable.append(row)

print(sourceTable[:10])

with open('resultX.json',"w", encoding='utf-8') as outputFile:
    json.dump(sourceTable, outputFile)
```
